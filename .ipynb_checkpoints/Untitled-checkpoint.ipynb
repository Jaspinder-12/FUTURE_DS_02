```python
# Missing values
df.isnull().sum()

# Blank spaces in TotalCharges
(df['TotalCharges'] == " ").sum()

# Convert TotalCharges
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check missing values again
df.isnull().sum()

# Remove missing rows
df.dropna(inplace=True)

# Check duplicates
df.duplicated().sum()

# Verify data types
df.info()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('ggplot')
```

```python
churn_counts = df['Churn'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(churn_counts.index, churn_counts.values)
plt.title('Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()

print(churn_counts)
```

```python
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Customers')
plt.show()
```

```python
contract = pd.crosstab(df['Contract'], df['Churn'])

contract.plot(kind='bar', figsize=(8,5))

plt.title('Contract Type vs Churn')
plt.xlabel('Contract Type')
plt.ylabel('Customers')
plt.show()
```

```python
internet = pd.crosstab(df['InternetService'], df['Churn'])

internet.plot(kind='bar', figsize=(8,5))

plt.title('Internet Service vs Churn')
plt.xlabel('Internet Service')
plt.ylabel('Customers')
plt.show()
```

```python
payment = pd.crosstab(df['PaymentMethod'], df['Churn'])

payment.plot(kind='bar', figsize=(10,5))

plt.title('Payment Method vs Churn')
plt.xlabel('Payment Method')
plt.ylabel('Customers')
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df[df['Churn']=='Yes']['MonthlyCharges'], bins=25, alpha=0.7, label='Churn')
plt.hist(df[df['Churn']=='No']['MonthlyCharges'], bins=25, alpha=0.7, label='Stayed')

plt.title('Monthly Charges Distribution')
plt.xlabel('Monthly Charges')
plt.ylabel('Customers')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df[df['Churn']=='Yes']['tenure'], bins=20, alpha=0.7, label='Churn')
plt.hist(df[df['Churn']=='No']['tenure'], bins=20, alpha=0.7, label='Stayed')

plt.title('Customer Tenure Distribution')
plt.xlabel('Months')
plt.ylabel('Customers')
plt.legend()

plt.show()
```

```python
senior = pd.crosstab(df['SeniorCitizen'], df['Churn'])

senior.plot(kind='bar', figsize=(6,4))

plt.title('Senior Citizen vs Churn')
plt.xlabel('Senior Citizen')
plt.ylabel('Customers')
plt.show()
```

```python
paperless = pd.crosstab(df['PaperlessBilling'], df['Churn'])

paperless.plot(kind='bar', figsize=(6,4))

plt.title('Paperless Billing vs Churn')
plt.xlabel('Paperless Billing')
plt.ylabel('Customers')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(6,5))
plt.imshow(numeric_df.corr(), cmap='coolwarm', interpolation='nearest')
plt.colorbar()

plt.xticks(range(len(numeric_df.columns)), numeric_df.columns, rotation=45)
plt.yticks(range(len(numeric_df.columns)), numeric_df.columns)

plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
